```python
cities_name=[]
countrycode=[]
for i in range(0,len(cities_df['lat'])):
    lat=cities_df.iloc[i]['lat']
    long=cities_df.iloc[i]["long"]
    city=citipy.nearest_city(lat,long)
    name=city.city_name
    cities_name.append(name)
    country=city.country_code
    countrycode.append(country)
    
cities_df['City']=cities_name
cities_df['Country']=countrycode
cities_df.head(10)
```

```python
cities_df.drop_duplicates("City", keep='first').reset_index(drop=True, inplace=True)
```

```python
cities_df["Max Temp"]=""
cities_df["Humidity"]= ""
cities_df["Cloud Cover"] = ""
cities_df["Wind Speed"] = ""
cities_df["Date"]=""
cities_df.head()
```

```python
 
for row in cities_df['City']:
    
    
    target_url="http://api.openweathermap.org/data/2.5/weather?q=%s&units=IMPERIAL&mode=json&APPID=%s" % ({row}, API_KEY)
    cities_weather = requests.get(target_url).json()
       
        
    try:
        cities_df["Max Temp"] = cities_weather["main"]["temp_max"]
        cities_df["Humidity"] = cities_weather["main"]["humidity"]
        cities_df["Cloud Cover"] = cities_weather["clouds"]["all"]
        cities_df["Wind Speed"] = cities_weather["wind"]["speed"]
        cities_df["Wind Speed"] = cities_weather["dt"]
        
    except:
        print("Missing city weather data.......skipping")
        
        
                                                                       
    
                    
# Turn columns to numeric because of empty rows
cities_df["Max Temp"] = pd.to_numeric(cities_df["Max Temp"], errors='coerce')
cities_df["Humidity"] = pd.to_numeric(cities_df["Humidity"], errors='coerce')
cities_df["Cloud Cover"] = pd.to_numeric(cities_df["Cloud Cover"], errors='coerce')
cities_df["Wind Speed"] = pd.to_numeric(cities_df["Wind Speed"], errors='coerce')

cities_df.head(10)
```
